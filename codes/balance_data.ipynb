```python
from fastai.vision import *
import numpy as np
import pandas as pd
from collections import Counter
from random import shuffle
import cv2
```

```python
train_data = np.load('D:\\DeepACS\\dataset\\training_data1.npy', allow_pickle=True)
directory = r'D:/DeepACS/dataset'
folder = r'None1'    # either A or D
path = Path(directory)
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
path.ls()
```

```python
classes = ['A', 'None', 'D']
folder_A = directory + '/' + 'A1/'
folder_D = directory + '/' + 'D1/'
folder_None = directory + '/' + 'None1/'
```

```python
lefts = []
rights = []
straights = []
shuffle(train_data)
```

```python
for data in train_data:
    image = data[0]
    choice = data[1]
    if choice[0] == 1:
        lefts.append([image, choice])
    if choice[2] == 1:
        rights.append([image, choice])
    if choice[1] == 1:
        straights.append([image, choice])

straights = straights[:len(rights)]
lefts = lefts[:len(rights)]
rights = rights[:len(rights)]

train_data = straights + lefts + rights
print(len(train_data))
print(len(straights))
print(len(lefts))
print(len(rights))
```

```python
i = 0
j = 0
k = 0
for data in train_data:
    image = data[0]
    choice = data[1]
    cv2.imshow('show', image)
    print(choice)
    
    if choice[0] == 1:
        cv2.imwrite(folder_A + str(i) + '.jpg', image)
        i += 1
    if choice[2] == 1:
        cv2.imwrite(folder_D + str(j) + '.jpg', image)
        j += 1
    if choice[1] == 1:
        cv2.imwrite(folder_None + str(k) + '.jpg', image)
        k += 1
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break
```
