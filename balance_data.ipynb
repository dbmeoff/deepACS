```python
from fastai.vision import *
import numpy as np
import pandas as pd
from collections import Counter
from random import shuffle
import cv2
```

```python
folders = ['A', 'D', 'None']
for i in folders:
    directory = r'dataset'
    folder = i    # either A or D
    path = Path(directory)
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)

path.ls()
```

```python
classes = ['A', 'None', 'D']
folder_A = directory + '/' + 'A/'
folder_D = directory + '/' + 'D/'
folder_None = directory + '/' + 'None/'
```

```python
training_data = []
for i in range(1, 4):
    train_data = np.load(r'dataset/training_data' + str(i) + '.npy', allow_pickle=True)
    lefts = []
    rights = []
    straights = []
    shuffle(train_data)
    for data in train_data:
        image = data[0]
        choice = data[1]
        if choice[0] == 1:
            lefts.append([image, choice])
        if choice[2] == 1:
            rights.append([image, choice])
        if choice[1] == 1:
            straights.append([image, choice])

    minimum = min(len(lefts), len(rights), len(straights))

    straights = straights[:minimum]
    lefts = lefts[:minimum]
    rights = rights[:minimum]

    train_data = straights + lefts + rights
    training_data.extend(train_data)
    print(len(train_data))
    print(len(straights))
    print(len(lefts))
    print(len(rights))
    print(len(training_data))
```

```python
i = 0
j = 0
k = 0
for data in training_data:
    image = data[0]
    choice = data[1]
    cv2.imshow('show', image)
    print(choice)
    
    if choice[0] == 1:
        cv2.imwrite(folder_A + str(i) + '.jpg', image)
        i += 1
    if choice[2] == 1:
        cv2.imwrite(folder_D + str(j) + '.jpg', image)
        j += 1
    if choice[1] == 1:
        cv2.imwrite(folder_None + str(k) + '.jpg', image)
        k += 1
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break
    print(i, j, k)
```
